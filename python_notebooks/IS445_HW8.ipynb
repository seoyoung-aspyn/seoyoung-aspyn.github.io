```python
import altair as alt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
building_url = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv'
buildings = pd.read_csv(building_url)

buildings.loc[buildings['Year Acquired']==0,'Year Acquired'] = np.nan
buildings.loc[buildings['Year Constructed']==0,'Year Constructed'] = np.nan
buildings.loc[buildings['Square Footage']==0,'Square Footage'] = np.nan
```

```python
alt.data_transformers.enable('json')
panzoom = alt.selection_interval(bind='scales')

building_inv_chart1 = alt.Chart(buildings).mark_bar().encode(
    x='Year Constructed:Q',
    y='Square Footage:Q',
    tooltip=['Year Constructed', 'Square Footage', 'Bldg Status']  # Adding tooltips here
).properties(
    width=800,
    height=300
).add_params(
    panzoom 
)

building_inv_chart2 = alt.Chart(buildings).mark_bar().encode(
    x=alt.X('Bldg Status:N', title='Building Status'),
    y=alt.Y('count(Bldg Status):Q', title='Count')
).properties(
    width=800,
    height=300
).add_params(
    panzoom  
)

building_charts = building_inv_chart1 | building_inv_chart2
building_charts
```
