```python
import altair as alt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
brush = alt.selection_interval(encodings=['x','y'])

chart1 = alt.Chart.from_dict({
    "data":{"url":'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv'},
    "mark":"bar",
    "width":600,
    "height":300,
    "encoding":{
    "x":{"field":"season", "type":"ordinal", "title":"Season"},
    "y":{"field":"visibility", "type":"quantitative", "title":"Visibility"},
  } 
}).add_selection(
    brush
)

chart2 = alt.Chart.from_dict({
    "data":{"url":'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv'},
    "mark":"bar",
    "width":600,
    "height":300,
    "encoding":{
    "x":{"field":"state", "type":"ordinal", "title":"State"},
    "y":{"field":"wind_speed", "type":"quantitative", "title":"Wind Speed"},
    }#// end encoding
}).transform_filter(
    brush
)

bfro_report = chart1 | chart2
bfro_report
```
