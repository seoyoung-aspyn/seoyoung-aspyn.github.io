```python
import altair as alt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
brush = alt.selection_interval(encodings=['x','y'])

chart1 = alt.Chart.from_dict({
    "data":{"url":'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv'},
    "mark":"rect",
    "width":300,
    "height":300,
    "encoding":{
    "x":{"bin":{"maxbins":20},"field":"uv_index", "type":"quantitative", "title":"UV Index"},
    "y":{"bin":{"maxbins":20}, "field":"visibility", "type":"quantitative", "title":"Visibility"},
    "color": {"field": "season", "type": "ordinal", "legend": {"title": "Season"}} 
  } 
}).add_selection(
    brush
)

chart2 = alt.Chart.from_dict({
    "data":{"url":'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv'},
    "mark":"rect",
    "width":600,
    "height":300,
    "encoding":{
    "x":{"field":"state", "type":"ordinal", "title":"State"},
    "y":{"bin":{"maxbins":60},"field":"wind_speed", "type":"quantitative", "title":"Wind Speed"},
    "color": {"field": "season", "type": "ordinal", "legend": {"title": "Season"}} 
    }
}).transform_filter(
    brush
)

bfro_reports = chart1 | chart2
bfro_reports
```
